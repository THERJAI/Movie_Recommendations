```python
import pandas as pd
```

```python
moives=pd.read_csv("tmdb_5000_credits.csv")
cradtis=pd.read_csv("tmdb_5000_movies.csv")
```

```python
cradtis.head(2)
```

```python
moives=moives.merge(cradtis,on="title")
```

```python
moives.head(2)
```

```python
moives.isnull().sum()
```

```python
moives = moives[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
moives.isnull().sum()
```

```python
moives.dropna(inplace=True)


```

```python
moives.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
moives.head()
```

```python
moives['genres'] = moives['genres'].apply(convert)
moives.head()
```

```python
moives['keywords'] = moives['keywords'].apply(convert)
moives.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
moives['cast'] = moives['cast'].apply(convert)
moives.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
moives['crew'] = moives['crew'].apply(fetch_director)
```

```python
moives.head()
```

```python
moives['overview'] = moives['overview'].apply(lambda x:x.split())
```

```python
moives['tags'] = moives['overview'] + moives['genres'] + moives['keywords'] + moives['cast'] + moives['crew']
```

```python
new = moives.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Superman')
```
